```
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

def printConfusionMatrix(pp, pn, np, nn):
    array = [[pp,pn],
            [np,nn]]

    df_cm = pd.DataFrame(array, ["Positif", "Négatif"], ["Positif", "Négatif"])
    # plt.figure(figsize=(10,7))
    sn.set(font_scale=1.2) # for label size
    sn.heatmap(df_cm, annot=True, annot_kws={"size": 16}, cmap="PuRd") # font size

    plt.show()
```

```
printConfusionMatrix(13, 2, 3, 2)
printConfusionMatrix(13, 2, 3, 2)
```

This notebook contain the confusion of matrix.

```
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

def printConfusionMatrix(title, vp, vn, fp, fn):
    array = [[vp,vn],
            [fp,fn]]

    df_cm = pd.DataFrame(array, ["Vrai", "Faux"], ["Positif", "Négatif"])
    # plt.figure(figsize=(10,7))
    sn.set(font_scale=1.2) # for label size
    sn.heatmap(df_cm, annot=True, annot_kws={"size": 16}, cmap="PuRd") # font size

    plt.title(title, fontsize=16)
    plt.show()
```

```
printConfusionMatrix("Matrice de confusion pour Watershed", 11, 0, 3, 9)
printConfusionMatrix("Matrice de confusion pour YOLO v5", 25, 0, 0, 0)
```
